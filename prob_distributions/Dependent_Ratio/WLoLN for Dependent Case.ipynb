```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def running_smpl_mean_plot(input_list,c):
    size = len(input_list)
    n = np.linspace(0,size-1,size)
    count = 0
    result_list = []
    while (count < size) :
        result_list.append(np.mean(input_list[0:count]))
        count += 1
##    plt.figure()
##    plt.plot(n,result_list,color="red")
##    plt.plot(n,c*((n+1)/(n+1)),color="blue")
##    plt.grid()
##    plt.show()
    return result_list
    
#parameters for P and Q
meanP = 10
sdP = .1
meanQ = 10
sdQ = .1
k = 0

N = 10000

result_arr = []

Total = 10
count1 = 0
while (count1 < Total):
    np.random.seed(100*count1)
    #Z=Q/(P-kQ)
    Pdata = np.random.normal(meanP,sdP,N)
    Qdata = np.random.normal(meanQ,sdQ,N)
    Zdata = Qdata / (Pdata - k * Qdata)
    c0 = -k/((k*k)+((sdP/sdQ)*(sdP/sdQ)))
    result = running_smpl_mean_plot(Zdata,c0)
    count1 += 1
    result_arr.append(result)
```

```python
plt.plot(result_arr[0])
plt.plot(result_arr[1])
plt.plot(result_arr[2])
plt.plot(result_arr[3])
plt.plot(result_arr[4])
plt.plot(result_arr[5])
plt.ylim(.999,1.001)
plt.ylabel("Value of Z")
plt.xlabel("Sample Size")
plt.title("Running Sample Means")
plt.show()
```
