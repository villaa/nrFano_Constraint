```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters for X and Y
meanX = 10
stdX = 1
meanY = 10
stdY = 1
N=100000

#Z=X/Y
Xdata = np.random.normal(meanX,stdX,N)
Ydata = np.random.normal(meanY,stdY,N)
Zdata = Xdata / Ydata

#Sample information
smpl_meanX = np.mean(Xdata)
smpl_meanY = np.mean(Ydata)
smpl_meanZ = np.mean(Zdata)
print("Sample Mean for X = ", smpl_meanX)
print("Sample Mean for Y = ", smpl_meanY)
print("Sample Mean for Z = ", smpl_meanZ)

#Switch between regular histogram and pdf
density = True

# Bins
bins = int(np.floor(3*np.cbrt(N)))
print("Bins = ", bins)

#Histograms
histX, bin_edgesX = np. histogram(Xdata, bins, density=density)
histY, bin_edgesY = np. histogram(Ydata, bins, density=density)
histZ, bin_edgesZ = np. histogram(Zdata, bins, density=density)

#Figure
plt.figure()
plt.plot(bin_edgesX[:-1], (histX), color="red", label = "x-plot")
plt.grid()
plt.legend()
plt.show()

plt.figure()
plt.plot(bin_edgesY[:-1], (histY), color="red", label = "y-plot")
plt.grid()
plt.legend()
plt.show()

plt.figure()
plt.plot(bin_edgesZ[:-1], (histZ), color="blue", label = "z-plot")
plt.grid()
plt.legend()
plt.show()
```
