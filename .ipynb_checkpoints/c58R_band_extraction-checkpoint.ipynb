# Using `uproot` to Get the Bands

The bands could in principle be extracted from the publication by being patient with DataThief, etc. I'm using that fact to be comfortable with simply getting them from the data file I used to construct Fig. 5 of the c58R paper. I stored all that data in a `ROOT` file and since `ROOT` is horrible, I'll use `uproot` to get them in Python format, here.

```python
import uproot
import numpy as np

yieldband = uproot.open('data/figuredata.root')['yieldBands402']
print(yieldband.keys())

nrmin = yieldband['nrMin']
nrmax = yieldband['nrMax']
ermin = yieldband['erMin']
ermax = yieldband['erMax']

cutdownfactor = 4
idx = np.arange(0,np.shape(nrmin._fX)[0])
nrminx = np.asarray(nrmin._fX)[idx%cutdownfactor==0]
nrminy = np.asarray(nrmin._fY)[idx%cutdownfactor==0]
idx = np.arange(0,np.shape(nrmax._fX)[0])
nrmaxx = np.asarray(nrmax._fX)[idx%cutdownfactor==0]
nrmaxy = np.asarray(nrmax._fY)[idx%cutdownfactor==0]

idx = np.arange(0,np.shape(ermin._fX)[0])
erminx = np.asarray(ermin._fX)[idx%cutdownfactor==0]
erminy = np.asarray(ermin._fY)[idx%cutdownfactor==0]
idx = np.arange(0,np.shape(ermax._fX)[0])
ermaxx = np.asarray(ermax._fX)[idx%cutdownfactor==0]
ermaxy = np.asarray(ermax._fY)[idx%cutdownfactor==0]

#print(nrmin._fX)
```

```python
#play around with some hits data stored in h5 file
#===============to suppress h5py warning see:
#https://github.com/h5py/h5py/issues/961
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import h5py
warnings.resetwarnings()
```

```python
#open and write file
of = h5py.File('c58R_bands.h5', 'w')


dset_hits = of.create_dataset("c58R_bands/nrmin_x", np.shape(nrminx), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = nrminx
dset_hits = of.create_dataset("c58R_bands/nrmin_y", np.shape(nrminy), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = nrminy


dset_hits = of.create_dataset("c58R_bands/nrmax_x", np.shape(nrmaxx), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = nrmaxx
dset_hits = of.create_dataset("c58R_bands/nrmax_y", np.shape(nrmaxy), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = nrmaxy

dset_hits = of.create_dataset("c58R_bands/ermin_x", np.shape(erminx), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = erminx
dset_hits = of.create_dataset("c58R_bands/ermin_y", np.shape(erminy), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = erminy

dset_hits = of.create_dataset("c58R_bands/ermax_x", np.shape(ermaxx), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = ermaxx
dset_hits = of.create_dataset("c58R_bands/ermax_y", np.shape(ermaxy), dtype=np.dtype('float64').type, compression="gzip", compression_opts=9)
dset_hits[...] = ermaxy

of.close()
```
